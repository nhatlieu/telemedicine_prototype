```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('./covid-19_symptoms_dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df['fever'] = (df['fever'] - 32) * (5/9)
df['fever'] = np.ceil(df['fever'] * 10) / 10
df.head()
```

```python
df['diffBreath'] = df['diffBreath'] + 1
df.head()
```

```python
df.describe()
```

```python
sc = StandardScaler()
df['fever'] = sc.fit_transform(df['fever'].array.reshape(-1, 1))
df['age'] = sc.fit_transform(df['age'].array.reshape(-1, 1))
df.head()
```
